```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

```

```python
# creating datasets X containing n_samples
# Y containing two classes
X, Y = make_blobs(n_samples=500, centers=2, random_state=0, cluster_std=0.40)
```

```python
X.shape, Y.shape
```

```python
# plt.scatter(X[:, 0], X[:, 1], c=Y, s=50, cmap='spring')
# plt.show()

# X[:, 0]: This selects the first column of the X array, which contains the x-axis values for the scatter plot.
# X[:, 1]: This selects the second column of the X array, which contains the y-axis values for the scatter plot.
# This sets the color of each point in the scatter plot based on the corresponding value in the Y array.
```

```python
# creating linspace between -1 to 3.5
xfit = np.linspace(-1, 3.5)

plt.scatter(X[:, 0], X[:, 1], c=Y, s=50, cmap='spring')

for m, b, d in [(1, 0.65, 0.33), (0.5, 1.6, 0.55), (-0.2, 2.9, 0.2)]:
    yfit = m * xfit + b
    plt.plot(xfit, yfit, '-k')
    plt.fill_between(xfit, yfit - d, yfit + d, edgecolor='none',
                     color='#AAAAAA', alpha=0.4)

plt.xlim(-1, 3.5);
plt.show()
```
